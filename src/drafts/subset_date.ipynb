```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv("crsp_small_sample.csv")

df.head()
```

```python
tot = df.shape[1]
num_assets = df.shape[1] // 5
```

```python
int_choices = np.random.choice(range(0, tot), num_assets, replace=False)
```

```python
len(int_choices)
```

```python
int_choices
```

```python
colnames = [df.columns[i] for i in int_choices]
df[["date"] + colnames].head()
```
